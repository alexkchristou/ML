```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('DATA/cancer_classification.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#df.rename(columns = {'test':'TEST'}, inplace = True) this renames a column named test to TEST
df.rename(columns = {'benign_0__mal_1':'label'}, inplace = True)
```

```python
sns.countplot('label',data=df)
```

```python
df.corr()['label'][:-1].sort_values().plot(kind='bar')
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(data=df.corr())
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
X = df.drop('label', axis=1).values
y = df['label'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train) #using the train set for scaling to avoid data leakage
```

```python
X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
X_train.shape #helps me check how many features I have on my training data set
```

```python
#setting up my neural network
model=Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu')) #reducing neurons by half on second layer
model.add(Dense(1,activation='sigmoid')) #output node activates as sigmoid for binary classification
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(x=X_train,y=y_train,epochs=600,verbose=1,validation_data=(X_test,y_test))
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop=EarlyStopping(monitor='val_loss',mode='min',verbose=1,patience=25)
```

```python
#Please rerun model before you fit again!
model.fit(x=X_train,y=y_train,epochs=600,verbose=1,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
#typical convergence/training visualisation sequence of the error function
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
#setting up my neural network - Adding dropout layers
model=Sequential()

model.add(Dense(30,activation='relu'))
model.add(Dense(15,activation='relu')) #reducing neurons by half on second layer
model.add(Dropout(0.5)) # dropout layer. rate is the percentage of neurons we are turning off at random.
model.add(Dense(1,activation='sigmoid')) #output node activates as sigmoid for binary classification
model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
#Please rerun model before you fit again!
model.fit(x=X_train,y=y_train,epochs=600,verbose=1,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
losses=pd.DataFrame(model.history.history)
losses.plot()
```

```python
y_pred=model.predict(X_test) 
predi=np.where(y_pred > 0.5, 1,0) #translate probabilities to 0 or 1) 
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
confusion_matrix(y_test,predi)
```

```python
print (classification_report(y_test,predi))
```
